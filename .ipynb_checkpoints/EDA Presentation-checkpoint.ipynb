import the necessary stuff

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
```

```python
# Read the data using a " " as the seperator (all data is in first column)
sns.set()
df = pd.read_csv("abalones.csv", delimiter = " ")
df.head()
```

```python
# Creating features VOLUME and Ratio
df['VOLUME'] = df['HEIGHT']*df['LENGTH']*df['DIAM']
df['RATIO'] = df['SHUCK']/df['VOLUME']

df.head()
```

```python
# Output summary statistics of the data
df.describe()
```

From looking at the summary table, I can assume that there are no missing values in the numeric columns of data due to the summary not displaying any minimums of 0. However to ensure that I could run np.where(pd.isnull(df) to display the index position of any null cells in the dataframe.

The pandas describe function shows us useful summary statistics for numerical variables. One important thing to note when analyzing the summary ouput, is that some of the variables (WHOLE, SHUCK, VOLUME) have MAX values that are significantly higher than their mean values. This could impact the distribution of those variables and might be represented as a skew in the data towards the right. We would need to identify these MAX values and look into if they are outliers, and then the method in which we would deal with them so that the data would be less skewed and more normal, leading to better model performance.

```python
# Using a crosstab to get the total value margins in the table
pd.crosstab(index=df['CLASS'], columns=df['SEX'], margins=True, margins_name = 'Total')
```

```python
# Creating grouped bar graph using a pivot table to format data correctly

df.pivot_table(index='CLASS', columns='SEX', aggfunc='size').plot(
    figsize=(15,7),
    kind='bar',
    xlabel= "Age Class",
    ylabel="Frequency",
    grid=True).set_title('Comparison Abalone Sex Frequencies', fontsize=18)
```

Doing some research on the variables we find out that CLASS is representive of the age group of the Abalone, where the age is defined as the number of rings multiplied by 1.5.

Given that insight on the variable we can see that majority of the sampled Abalones are in A3 class. This is consistent with both the male and female abalones, but not with the infants as they were mostly determined to be in the A1 and A2 class due to them being newborns. This makes sense as the newborn infants are more prevalent in the younger classes wheras the adults who have lived longed and have had their sex determined are in the A3 class.

In the graph above we can see that following class A3 there is a large drop in the number of abalones across all sexes. We can assume then that the A3 class is where an abalone is generally fully grown and reproducing and this could be part of why their population drops off in the following classes. Looking back at the summary statistics we can see that the max value for rings on an abalone in the dataset is 25. Therefore that abalone was 37.5 years old. Using this and mean of 11 rings, we can have a general estimation of the life expectancy of an abalone. 

It is interesting to note that as infants grow older their change in age and from being classed as an infant to a male or a female is represented in the graph by their decreasing frequency after CLASS A3.

Inversely, we can see that it is very hard to determine the sex of an immature abalone. We see that out of 108 total abalones in the immature A1 class, only 4.63% (5) were classified as female and 11.11% (12) as male. The rest 84.26% are infants (91). This could be due to difficulty making out the features of the infant abalone, as they are very little and haven't developed. I understand that this was a significant problem when using the rings as age predictors, so it would make sense for the same problem to be present with the sex.

There are also trends within the sex groups. One example is the sharp decrease in the amount of infants from class A2 to A3. From this we can assume that by CLASS A3, the reproductive features that are required to determine the sex are developed. When looking at the amount of females after the class A3, this makes sense as after reproduction animals are vulnerable and this could also be represented by their decreasing population in CLASS A4, A5 and A6 as they are dying out after the reproduce.

Finally there are some infants that are in the oldest growth category, there are clearly not infants but also are abalones that lack the features used to determine their sex. We can determine from this that they might not all be able to reproduce or that some of the abalones can develop these features at a much later date.

```python
# Selecting a random sample of 200 observations as 'work'

work = df.sample(n=200,random_state=123)
work.shape
```

```python
# Create Scatter Matrix
pd.plotting.scatter_matrix(frame=work[['LENGTH','DIAM','HEIGHT','WHOLE','SHUCK','RINGS']], figsize=(15,12))
```

Looking at the matrix above, we can see a lot positive coorelation. We see the strongest positive coorelation between the features LENGTH vs DIAM. This makes sense as it is natural for an abalone with a large diameter to have a have a longer shell length. WHOLE and SHUCK are also strongly positively coorelated. We can see this follows a very straight linear line, however towards the end there seems to be some variance, although still following the general linear trend. This makes perfect sense since the SHUCK (weight of the shucked meat) is a part of the WHOLE (weight of the whole abalone).

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=df, y='VOLUME', x='WHOLE', hue='CLASS', alpha=0.6).set_title('Scatter Plot of Volume versus Whole', fontsize = 18)

# Get Max Ratio
df3 = df.copy()
df3['ratio2'] = df['SHUCK']/df['WHOLE']
max_ratio = df3['ratio2'].max()
max_ratio
df3.drop('ratio2', axis=1)

# Plotting WHOLE versus VOLUME using the start point as 0,0 and the slope as the max_ratio
plt.figure(figsize=(10,5))
ax = sns.scatterplot(data=df, x='WHOLE', y='SHUCK', hue='CLASS', alpha=0.6)
ax.axline((0,0), slope=max_ratio)
```

In the plot of VOLUME vs WHOLE, as WHOLE increases, VOLUME tends to increase as well. This is similar to the SHUCK vs WHOLE plot, in which as WHOLE increases, SHUCK increases. From this we can see that both the graphs are similar in that they both have strong positive relationships between the features. Therefore we know that WHOLE is a good predictor of both the VOLUME and the SHUCK of an abalone. This is not surprising since as the total weight of an abalone increases, we would generally expect for it's volume and the weight of it's shucked meat to increase as well.

In the plot of SHUCK vs WHOLE, there is less variability. In the SHUCK vs WHOLE plot, we can see that the A4 group of abalones are showing large WHOLE weights with less shucked meat weight. We can then assume that most of their weight is coming from their shell now and they do not have as much meat despite being older when compared to the A3 classes, which represent younger abalones that have similar WHOLE weight, but more of it comes from shucked meat weight.

This is shown on the SHUCK vs WHOLE graph with the young abalones being closer to where the trendline would be and the older abalones being below it.





```python
# Histograms
df.hist(column='RATIO', by='SEX', figsize=(15,5), color = 'green', layout=(1,3))


# Boxplot
bplot = df.boxplot(
    column='RATIO',
    by='SEX',
    figsize=(15,6),
    color={'boxes':'blue', 'medians':'white'},
    showmeans=True,
    patch_artist=True)

plt.suptitle('') # Get rid of default generic title
bplot.set_title('Ratio Differentiated by Sex', fontsize=18)
bplot.set_ylabel('Ratio', fontsize=18)
bplot.set_xlabel('Sex', fontsize=18)


# QQ Plots
fig, axes = plt.subplots(1,3,figsize=(14,5))
F,I,M = df[df['SEX'] == 'F'], df[df['SEX'] == 'I'], df[df['SEX'] == 'M']
F,I,M = F[['RATIO']], I[['RATIO']], M[['RATIO']]
arrF, arrI, arrM = F.values.flatten(), I.values.flatten(), M.values.flatten() # Need to flatten data since it only takes 1D arrays
F = sm.qqplot(arrF, line='s', ax=axes[0])
I = sm.qqplot(arrI, line='s',ax=axes[1])
M = sm.qqplot(arrM, line='s',ax=axes[2])
plt.tight_layout()
plt.show()
```

Histograms:
The histograms report a general normal relationship of the data as it follows a bell curve shape. The histograms are ever so slightly skewed to the right, and this could be corrected with a transform such as normalising the data.

Boxplots:
When looking at the boxplots, we can see that there are outliers. For the female category there are outliers that are too high and too low. For the infant and the male category, there are only outliers that are too high.

Looking at the female outliers we can see that there is an outlier that has a value greater than 0.30 for the ratio. This is significant because that is more than twice the value of the mean ratio for abalones (0.14), so it will be important to take a look at dealing with. This outlier is also easily visible on the QQ plot as it does not follow the general trend of the data.

QQ Plot:
Looking at the QQ plot we can see that the female QQ plot is the closest to being normal. There is not much variation outside of the outliers in the three groups. At a glance it appears that the infant category has the most values that deviate away from the normal line.

I do not think that these variables will be good predictors for age. Looking at the box and whisker graph we can see that there is a lot of overlap between the different boxes. This overlap means that an abalone of 400cm^3 could belong to A3, A4, A5 or A6.

There is also overlap with the WHOLE vs RINGS boxplot. An abalone that has a whole weight of 130 grams could be part of A3, A4 or A5. However just like the VOLUME vs RINGS boxplot, there is less overlap with the classes A1 and A2.

The scatterplots show that around the CLASS A3, the abalones are generally mature.

I do not think that we can be condifent about these predictors for the general population of abalones or the sample that we have, but perhaps for a subset of younger, less heavy abalones they would make good predictors.
